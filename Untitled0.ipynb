```python
# Importing tools (libraries) for our AI model
from sklearn.linear_model import LogisticRegression

# Training data: [ages]
X = [[5], [12], [25], [40], [60]]

# Labels: 1 = Likes ice cream, 0 = Doesn't like
y = [1, 1, 1, 0, 0]

# Creating the model
model = LogisticRegression()

# Training the model with data
model.fit(X, y)

# Making a prediction
test_ages = [[8], [30], [45], [70]]

for age in test_ages:
    result = model.predict([age])
    print(f"Prediction for age {age[0]}:", "Likes Ice Cream üç¶" if result[0] == 1 else "Doesn't Like üòÖ")

import matplotlib.pyplot as plt
import numpy as np

# Ages from 0 to 80
ages = np.arange(0, 81).reshape(-1, 1)

# Model's prediction probabilities
probabilities = model.predict_proba(ages)[:, 1]  # chance of 'likes ice cream'

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(ages, probabilities, label="Probability of Liking Ice Cream")
plt.scatter([5, 12, 25, 40, 60], [1, 1, 1, 0, 0], color='red', label="Training Data")
plt.xlabel("Age")
plt.ylabel("Probability")
plt.title("AI's Ice Cream Prediction Curve")
plt.legend()
plt.grid(True)
plt.show()

```
